# EDA Alternative fuel Vehicles Us

```python
import pandas as pd

# Load the dataset to analyze
file_path = 'Alternative Fuel Vehicles US.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns

# Función para calcular el porcentaje de valores faltantes
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_values / len(df)) * 100

# Identificar outliers en las variables numéricas
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
outliers = {col: df[col][(df[col] < df[col].quantile(0.05)) | (df[col] > df[col].quantile(0.95))]
            for col in numeric_columns}

# Graficar los valores faltantes como un gráfico de barras
plt.figure(figsize=(12, 6))
missing_percentage.plot(kind='bar', color='teal', alpha=0.7)
plt.title('Porcentaje de Valores Faltantes por Columna', fontsize=16)
plt.ylabel('Porcentaje', fontsize=12)
plt.xlabel('Columnas', fontsize=12)
plt.xticks(rotation=90)
plt.show()

# Graficar diagramas de caja (boxplots) para los outliers numéricos
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x=col, color='orange')
    plt.title(f'Diagrama de Caja para {col}', fontsize=14)
    plt.show()

# Visualizar las columnas categóricas con gráficos de barras (las primeras 6 por claridad)
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns[:6]:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Gráfico de Frecuencias para {col}', fontsize=14)
    plt.show()

# Matriz de correlación para las variables numéricas
correlation_matrix = df[numeric_columns].corr()

# Mapa de calor (heatmap) para las correlaciones
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Mapa de Calor de Correlaciones para Variables Numéricas', fontsize=16)
plt.show()

# Preparar el resumen de valores faltantes y outliers
missing_values_df = pd.DataFrame(missing_percentage, columns=["Porcentaje de Valores Faltantes"])
outliers_summary_df = pd.DataFrame({col: [len(outliers[col])] for col in outliers}, index=["Cantidad de Outliers"]).T

missing_values_df, outliers_summary_df
```

```python
# Ajustar los gráficos de frecuencia para 'Model' y 'Manufacturer' para mostrar los 15 más frecuentes (asi no satura el grafico)

# Top 15 modelos
top_models = df['Model'].value_counts().head(15)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_models.index, x=top_models.values, palette='viridis')
plt.title('Top 15 Modelos por Frecuencia', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)
plt.ylabel('Modelo', fontsize=12)
plt.show()

# Top 15 fabricantes
top_manufacturers = df['Manufacturer'].value_counts().head(15)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_manufacturers.index, x=top_manufacturers.values, palette='viridis')
plt.title('Top 15 Fabricantes por Frecuencia', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)
plt.ylabel('Fabricante', fontsize=12)
plt.show()

```

# Relaciones entre variables
1. Autonomía eléctrica vs. Economía de combustible alternativa (ciudad)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='All-Electric Range', y='Alternative Fuel Economy City')
plt.title('Autonomía Eléctrica vs. Economía de Combustible (Ciudad)', fontsize=14)
plt.xlabel('Autonomía Eléctrica (millas)', fontsize=12)
plt.ylabel('Economía de Combustible (mpg)', fontsize=12)
plt.show()
```

 Se espera observar que los vehículos con mayor autonomía eléctrica también presenten una mejor economía de combustible en ciudad, especialmente en modelos híbridos enchufables. Si hay dispersión significativa, puede indicar que otros factores influyen en la eficiencia.

2. Fabricante vs. Economía de combustible combinada

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Manufacturer', y='Conventional Fuel Economy Combined')
plt.xticks(rotation=90)
plt.title('Fabricante vs. Economía de Combustible Combinada', fontsize=14)
plt.xlabel('Fabricante', fontsize=12)
plt.ylabel('Economía de Combustible (mpg)', fontsize=12)
plt.show()
```

Los fabricantes con tecnología más avanzada, como Tesla o Toyota, deberían mostrar valores más altos de eficiencia. Las diferencias significativas entre fabricantes pueden destacar enfoques tecnológicos variados.

3. Categoría del vehículo vs. Economía de combustible combinada

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Conventional Fuel Economy Combined')
plt.xticks(rotation=45)
plt.title('Categoría del Vehículo vs. Economía de Combustible Combinada', fontsize=14)
plt.xlabel('Categoría del Vehículo', fontsize=12)
plt.ylabel('Economía de Combustible (mpg)', fontsize=12)
plt.show()
```

Se espera que vehículos más pequeños, como sedanes, tengan una mayor economía de combustible, mientras que SUVs y camiones presenten menor eficiencia debido a su peso y tamaño.

4. Cilindrada del motor vs. Economía de combustible combinada

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Engine Cylinder Count', y='Conventional Fuel Economy Combined')
plt.title('Cilindros del Motor vs. Economía de Combustible', fontsize=14)
plt.xlabel('Número de Cilindros', fontsize=12)
plt.ylabel('Economía de Combustible (mpg)', fontsize=12)
plt.show()
```

Se espera una relación inversa: a medida que aumenta el número de cilindros, la economía de combustible disminuye, reflejando el mayor consumo energético de motores grandes.

5. Año del modelo vs. Eficiencia del combustible

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Model Year', y='Conventional Fuel Economy Combined')
plt.title('Año del Modelo vs. Economía de Combustible', fontsize=14)
plt.xlabel('Año del Modelo', fontsize=12)
plt.ylabel('Economía de Combustible (mpg)', fontsize=12)
plt.show()
```

Los modelos más recientes deberían mostrar una mejora en la eficiencia del combustible debido a avances tecnológicos y regulaciones más estrictas en emisiones.



6. Sistema de propulsión vs. Economía de combustible

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Fuel', y='Alternative Fuel Economy Combined')
plt.xticks(rotation=90)
plt.title('Sistema de Propulsión vs. Economía de Combustible', fontsize=14)
plt.xlabel('Sistema de Propulsión', fontsize=12)
plt.ylabel('Economía de Combustible (mpg)', fontsize=12)
plt.show()
```

Los vehículos eléctricos deberían mostrar la mejor economía de combustible, seguidos de los híbridos. Los vehículos de combustibles alternativos podrían presentar variabilidad según la tecnología utilizada.

7. Rango total PHEV vs. Rango eléctrico

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PHEV Total Range', y='All-Electric Range')
plt.title('Rango Total PHEV vs. Rango Eléctrico', fontsize=14)
plt.xlabel('Rango Total PHEV (millas)', fontsize=12)
plt.ylabel('Rango Eléctrico (millas)', fontsize=12)
plt.show()
```

Se espera una relación positiva entre el rango eléctrico y el rango total en vehículos híbridos enchufables, aunque otros factores (como capacidad de la batería y tipo de conducción) podrían generar dispersión.

8. Fabricante vs. Número de pasajeros

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Manufacturer', y='Number of Passengers')
plt.xticks(rotation=90)
plt.title('Fabricante vs. Número de Pasajeros', fontsize=14)
plt.xlabel('Fabricante', fontsize=12)
plt.ylabel('Número de Pasajeros', fontsize=12)
plt.show()
```

Fabricantes como Ford y Chevrolet, que producen vehículos comerciales y familiares, deberían mostrar una mayor capacidad promedio de pasajeros en comparación con fabricantes de lujo como BMW o Porsche.
