# EDA Alternative fuel Vehicles Us

```python
import pandas as pd

# Load the dataset to analyze
file_path = 'Alternative Fuel Vehicles US.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns

# FunciÃ³n para calcular el porcentaje de valores faltantes
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_values / len(df)) * 100

# Identificar outliers en las variables numÃ©ricas
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
outliers = {col: df[col][(df[col] < df[col].quantile(0.05)) | (df[col] > df[col].quantile(0.95))]
            for col in numeric_columns}

# Graficar los valores faltantes como un grÃ¡fico de barras
plt.figure(figsize=(12, 6))
missing_percentage.plot(kind='bar', color='teal', alpha=0.7)
plt.title('Porcentaje de Valores Faltantes por Columna', fontsize=16)
plt.ylabel('Porcentaje', fontsize=12)
plt.xlabel('Columnas', fontsize=12)
plt.xticks(rotation=90)
plt.show()

# Graficar diagramas de caja (boxplots) para los outliers numÃ©ricos
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x=col, color='orange')
    plt.title(f'Diagrama de Caja para {col}', fontsize=14)
    plt.show()

# Visualizar las columnas categÃ³ricas con grÃ¡ficos de barras (las primeras 6 por claridad)
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns[:6]:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'GrÃ¡fico de Frecuencias para {col}', fontsize=14)
    plt.show()

# Matriz de correlaciÃ³n para las variables numÃ©ricas
correlation_matrix = df[numeric_columns].corr()

# Mapa de calor (heatmap) para las correlaciones
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Mapa de Calor de Correlaciones para Variables NumÃ©ricas', fontsize=16)
plt.show()

# Preparar el resumen de valores faltantes y outliers
missing_values_df = pd.DataFrame(missing_percentage, columns=["Porcentaje de Valores Faltantes"])
outliers_summary_df = pd.DataFrame({col: [len(outliers[col])] for col in outliers}, index=["Cantidad de Outliers"]).T

missing_values_df, outliers_summary_df
```

```python
# Ajustar los grÃ¡ficos de frecuencia para 'Model' y 'Manufacturer' para mostrar los 15 mÃ¡s frecuentes (asi no satura el grafico)

# Top 15 modelos
top_models = df['Model'].value_counts().head(15)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_models.index, x=top_models.values, palette='viridis')
plt.title('Top 15 Modelos por Frecuencia', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)
plt.ylabel('Modelo', fontsize=12)
plt.show()

# Top 15 fabricantes
top_manufacturers = df['Manufacturer'].value_counts().head(15)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_manufacturers.index, x=top_manufacturers.values, palette='viridis')
plt.title('Top 15 Fabricantes por Frecuencia', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)
plt.ylabel('Fabricante', fontsize=12)
plt.show()

```

# Relaciones entre variables
1. AutonomÃ­a elÃ©ctrica vs. EconomÃ­a de combustible alternativa (ciudad)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='All-Electric Range', y='Alternative Fuel Economy City')
plt.title('AutonomÃ­a ElÃ©ctrica vs. EconomÃ­a de Combustible (Ciudad)', fontsize=14)
plt.xlabel('AutonomÃ­a ElÃ©ctrica (millas)', fontsize=12)
plt.ylabel('EconomÃ­a de Combustible (mpg)', fontsize=12)
plt.show()
```

 Se espera observar que los vehÃ­culos con mayor autonomÃ­a elÃ©ctrica tambiÃ©n presenten una mejor economÃ­a de combustible en ciudad, especialmente en modelos hÃ­bridos enchufables. Si hay dispersiÃ³n significativa, puede indicar que otros factores influyen en la eficiencia.

2. Fabricante vs. EconomÃ­a de combustible combinada

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Manufacturer', y='Conventional Fuel Economy Combined')
plt.xticks(rotation=90)
plt.title('Fabricante vs. EconomÃ­a de Combustible Combinada', fontsize=14)
plt.xlabel('Fabricante', fontsize=12)
plt.ylabel('EconomÃ­a de Combustible (mpg)', fontsize=12)
plt.show()
```

Los fabricantes con tecnologÃ­a mÃ¡s avanzada, como Tesla o Toyota, deberÃ­an mostrar valores mÃ¡s altos de eficiencia. Las diferencias significativas entre fabricantes pueden destacar enfoques tecnolÃ³gicos variados.

3. CategorÃ­a del vehÃ­culo vs. EconomÃ­a de combustible combinada

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Conventional Fuel Economy Combined')
plt.xticks(rotation=45)
plt.title('CategorÃ­a del VehÃ­culo vs. EconomÃ­a de Combustible Combinada', fontsize=14)
plt.xlabel('CategorÃ­a del VehÃ­culo', fontsize=12)
plt.ylabel('EconomÃ­a de Combustible (mpg)', fontsize=12)
plt.show()
```

Se espera que vehÃ­culos mÃ¡s pequeÃ±os, como sedanes, tengan una mayor economÃ­a de combustible, mientras que SUVs y camiones presenten menor eficiencia debido a su peso y tamaÃ±o.

4. Cilindrada del motor vs. EconomÃ­a de combustible combinada

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Engine Cylinder Count', y='Conventional Fuel Economy Combined')
plt.title('Cilindros del Motor vs. EconomÃ­a de Combustible', fontsize=14)
plt.xlabel('NÃºmero de Cilindros', fontsize=12)
plt.ylabel('EconomÃ­a de Combustible (mpg)', fontsize=12)
plt.show()
```

Se espera una relaciÃ³n inversa: a medida que aumenta el nÃºmero de cilindros, la economÃ­a de combustible disminuye, reflejando el mayor consumo energÃ©tico de motores grandes.

5. AÃ±o del modelo vs. Eficiencia del combustible

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Model Year', y='Conventional Fuel Economy Combined')
plt.title('AÃ±o del Modelo vs. EconomÃ­a de Combustible', fontsize=14)
plt.xlabel('AÃ±o del Modelo', fontsize=12)
plt.ylabel('EconomÃ­a de Combustible (mpg)', fontsize=12)
plt.show()
```

Los modelos mÃ¡s recientes deberÃ­an mostrar una mejora en la eficiencia del combustible debido a avances tecnolÃ³gicos y regulaciones mÃ¡s estrictas en emisiones.



6. Sistema de propulsiÃ³n vs. EconomÃ­a de combustible

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Fuel', y='Alternative Fuel Economy Combined')
plt.xticks(rotation=90)
plt.title('Sistema de PropulsiÃ³n vs. EconomÃ­a de Combustible', fontsize=14)
plt.xlabel('Sistema de PropulsiÃ³n', fontsize=12)
plt.ylabel('EconomÃ­a de Combustible (mpg)', fontsize=12)
plt.show()
```

Los vehÃ­culos elÃ©ctricos deberÃ­an mostrar la mejor economÃ­a de combustible, seguidos de los hÃ­bridos. Los vehÃ­culos de combustibles alternativos podrÃ­an presentar variabilidad segÃºn la tecnologÃ­a utilizada.

7. Rango total PHEV vs. Rango elÃ©ctrico

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PHEV Total Range', y='All-Electric Range')
plt.title('Rango Total PHEV vs. Rango ElÃ©ctrico', fontsize=14)
plt.xlabel('Rango Total PHEV (millas)', fontsize=12)
plt.ylabel('Rango ElÃ©ctrico (millas)', fontsize=12)
plt.show()
```

Se espera una relaciÃ³n positiva entre el rango elÃ©ctrico y el rango total en vehÃ­culos hÃ­bridos enchufables, aunque otros factores (como capacidad de la baterÃ­a y tipo de conducciÃ³n) podrÃ­an generar dispersiÃ³n.

8. Fabricante vs. NÃºmero de pasajeros

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Manufacturer', y='Number of Passengers')
plt.xticks(rotation=90)
plt.title('Fabricante vs. NÃºmero de Pasajeros', fontsize=14)
plt.xlabel('Fabricante', fontsize=12)
plt.ylabel('NÃºmero de Pasajeros', fontsize=12)
plt.show()
```
