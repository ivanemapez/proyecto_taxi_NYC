3. Analisis de variables categoricas en frecuencias de valores y relaciones potenciales. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo
file_path = 'ElectricCarData_Norm.csv'  # Cambia esto por la ruta de tu archivo
df = pd.read_csv(file_path)

# Conversión de columnas relevantes a numéricas (si no se ha hecho antes)
def clean_numeric_column(column, remove_text=[' sec', ' km/h', ' km', ' Wh/km', ' km/h']):
    for text in remove_text:
        column = column.str.replace(text, '', regex=False)
    return pd.to_numeric(column, errors='coerce')

df['Accel'] = clean_numeric_column(df['Accel'], [' sec'])
df['TopSpeed'] = clean_numeric_column(df['TopSpeed'], [' km/h'])
df['Range'] = clean_numeric_column(df['Range'], [' km'])
df['Efficiency'] = clean_numeric_column(df['Efficiency'], [' Wh/km'])
df['FastCharge'] = clean_numeric_column(df['FastCharge'], [' km/h'])

# Variables categóricas
categorical_columns = ['Brand', 'Model', 'RapidCharge', 'PowerTrain', 'PlugType', 'BodyStyle', 'Segment']

# Frecuencia de valores categóricos
categorical_analysis = {col: df[col].value_counts() for col in categorical_columns}

# Paleta de colores
palette = sns.color_palette("viridis")

# Graficar distribuciones de valores categóricos
for col, analysis in categorical_analysis.items():
    plt.figure(figsize=(10, 5))
    sns.barplot(x=analysis.index, y=analysis.values, palette=palette)
    plt.title(f"Frecuencia de valores en {col}")
    plt.ylabel("Frecuencia")
    plt.xlabel(col)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Relación entre variables categóricas y PriceEuro
for col in ['RapidCharge', 'PowerTrain', 'BodyStyle', 'Segment']:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=col, y='PriceEuro', data=df, palette=palette)
    plt.title(f"Relación entre {col} y PriceEuro")
    plt.ylabel("Precio (€)")
    plt.xlabel(col)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Matriz de correlación para variables numéricas
numerical_columns = ['Accel', 'TopSpeed', 'Range', 'Efficiency', 'FastCharge', 'Seats', 'PriceEuro']
correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()

# Histogramas de variables numéricas
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=15, color=palette[0])
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()

```

1. Frecuencia de Valores Categóricos
    - Brand: Tesla domina el mercado con la mayor cantidad de modelos eléctricos, seguido por Audi, Nissan y Volkswagen. Esto refleja una estrategia agresiva de Tesla y un interés notable de otras marcas importantes.
    - RapidCharge: La mayoría de los vehículos (98 de 103) tienen carga rápida disponible, indicando que es una característica estándar en los autos eléctricos modernos.
    - PowerTrain: Los vehículos con tracción total (All Wheel Drive) son los más comunes, lo que sugiere una preferencia por rendimiento y seguridad.
    - BodyStyle: Los SUV son el tipo de carrocería más popular, seguido de los hatchbacks. Esto muestra la preferencia de los consumidores por vehículos espaciosos y prácticos.
    - Segment: Los autos eléctricos están mayormente en los segmentos C y B (compactos y subcompactos), lo que refleja un enfoque en la accesibilidad y el uso urbano.

2. Relaciones Categóricas-Numéricas

1. Relación entre Eficiencia y Autonomía (Efficiency vs. Range)

```python
# 1. Relación entre Eficiencia y Autonomía (Efficiency vs. Range)
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Efficiency', y='Range', hue='Brand', palette='viridis', legend=True)
for i in range(len(df)):
    plt.text(df['Efficiency'][i] + 1, df['Range'][i], df['Model'][i], fontsize=9, alpha=0.7)
plt.title("Relación entre Eficiencia y Autonomía por Modelo")
plt.xlabel("Eficiencia (Wh/km)")
plt.ylabel("Autonomía (km)")
plt.tight_layout()
plt.show()
```

Relación entre Eficiencia y Autonomía por Modelo
El gráfico revela que los modelos más eficientes (con menor consumo en Wh/km) tienden a tener autonomías mayores, aunque no siempre es una relación directa. Algunos puntos clave:

- Modelos como el One y el Tesla Model 3 destacan por su combinación de alta eficiencia y autonomía, mostrando que las marcas están optimizando su tecnología para maximizar el rendimiento por carga. 
- En contraste, modelos menos eficientes como el Audi e-tron ofrecen autonomías competitivas, pero a costa de un mayor consumo energético, probablemente debido a su tamaño, peso y características de lujo.
- Esto sugiere que las decisiones de diseño, como el tamaño de la batería y la aerodinámica, tienen un impacto significativo en la relación entre eficiencia y autonomía.

La gráfica destaca cómo la eficiencia energética puede no ser el único factor determinante en el rango, dado que vehículos menos eficientes aún logran ofrecer distancias significativas con una mayor capacidad de batería.

2. Relación entre Precio y Autonomía (PriceEuro vs. Range)

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='PriceEuro', y='Range', hue='Brand', palette='magma', legend=True)
for i in range(len(df)):
    plt.text(df['PriceEuro'][i] + 500, df['Range'][i], df['Model'][i], fontsize=9, alpha=0.7)
plt.title("Relación entre Precio y Autonomía por Modelo")
plt.xlabel("Precio (€)")
plt.ylabel("Autonomía (km)")
plt.tight_layout()
plt.show()
```

Relación entre Precio y Autonomía por Modelo
En este gráfico, se observa una relación positiva entre el precio y la autonomía, lo que indica que los modelos más caros suelen ofrecer mayores rangos. Sin embargo, hay excepciones notables:

- Los modelos de lujo, como el Lucid Air, combinan un precio elevado con autonomías excepcionales, destacándose en la parte superior derecha del gráfico.
- En el rango medio, marcas como Hyundai y Tesla ofrecen una buena relación entre precio y autonomía, posicionándose como opciones equilibradas para los consumidores.
- Los modelos más económicos, como el Renault Zoe, tienen autonomías limitadas, pero son ideales para usuarios urbanos que priorizan el costo sobre la distancia.

Este gráfico sugiere que el rango es un factor diferenciador importante en el precio, especialmente en los segmentos premium, donde los consumidores están dispuestos a pagar más por vehículos con mayor autonomía y tecnologías avanzadas.
