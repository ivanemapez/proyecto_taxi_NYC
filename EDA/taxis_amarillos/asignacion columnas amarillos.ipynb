```python
datas =[data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11,data12,data13,data14,data15,data16,data17,
        data18,data19,data20,data21,data22,data23,data24,data25,data26,data27,data28,data29,data30,data31,data32,data33]
```

```python
columnas_sel =['passenger_count','trip_distance','payment_type','day_pickup','hour_pickup','year','mes','duracion del viaje','tarifa - peajes']
```

```python
i=1
for data in datas:



    # garantizar que la columna este en formato datetime
    data['tpep_pickup_datetime'] = pd.to_datetime(data['tpep_pickup_datetime'])


    # Crear la columna para el día
    data['day_pickup'] = data['tpep_pickup_datetime'].dt.day

    # Crear la columna para la hora
    data['hour_pickup'] = data['tpep_pickup_datetime'].dt.hour

    # Crear la columna para el ano
    data['year'] = data['tpep_pickup_datetime'].dt.year

    # Crear la columna para el mes
    data['mes'] = data['tpep_pickup_datetime'].dt.month

    # Crear la columna duracion de viaje

    data['duracion del viaje'] = data['tpep_dropoff_datetime'] -  data['tpep_pickup_datetime']

    # Convertir la columna 'duracion del viaje' a formato timedelta
    data['duracion del viaje'] = pd.to_timedelta(data['duracion del viaje'])

    # Convertir la duración a minutos totales
    data['duracion del viaje'] = data['duracion del viaje'].dt.total_seconds() / 60

    # Crear columna tarifa - peaje

    data['tarifa - peajes']= data['total_amount'] -  data['tolls_amount']

    # Tomar el 20% del dataset como muestra
    datas[i-1] = data.sample(frac=0.2, random_state=42)

    print(f'data{i} iterada')
    i=i+1


```

```python
for i in range(len(datas)):
    datas[i] = datas[i][columnas_sel]
    print(f'columna definida en data{i+1}')
```

```python
data1,data2,data3,data4,data5,data6,data7 = datas[:7]

data8,data9,data10,data11,data12,data13,data14,data15,data16,data17,data18,data19,data20,data21,data22,data23,data24,data25,data26,data27,data28,data29,data30,data31,data32,data33= datas[7:]
```

```python
# Verificamos que los datasets tengan las mismas columnas
if list(data1.columns) == list(data2.columns) == list(data3.columns)==list(data4.columns)==list(data5.columns)==list(data16.columns)==list(data7.columns)==list(data8.columns)==list(data9.columns)==list(data10
.columns)==list(data11.columns)==list(data12.columns)==list(data13.columns)==list(data14.columns)==list(data15.columns)==list(data16.columns)==list(data17.columns)==list(data18.columns)==list(data19.columns)==list(data20.columns)==list(data21.columns)==list(data22
.columns)==list(data23.columns)==list(data24.columns)==list(data25.columns)==list(data26.columns)==list(data27.columns)==list(data28.columns)==list(data29.columns)==list(data30.columns)==list(data31.columns)==list(data32.columns)==list(data33.columns):
    # Unifica los datasets
    viajes_taxis_y= pd.concat([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11,data12,data13,data14,data15,data16,data17,data18,data19,data20,data21,data22,data23,data24,data25,data26,data27,data28,data29,data30,data31,data32,data33], ignore_index=True)
    print("Datasets unificados con éxito.")
else:
    print("Los datasets no tienen las mismas columnas y no se pueden unificar.")
```

```python
# Guarda el dataset unificado en un nuevo archivo parquet
viajes_taxis_y.to_parquet("viajes_taxis_y.parquet", engine = 'pyarrow',index=False)
```
